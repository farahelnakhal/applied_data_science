## Part II: Retrieving Selected Columns

From the dataset you just read from the csv file as a DataFrame, retrieve only the columns that represent grades for problem sets (i.e. *P1, P2, P3, ... P10*)

```python
df.filter(regex="^P[0-9]+$") #filter using regex

#^: start of string (no characters before)
#P: the prefix for PS cols
#[0-9]: one or more digits (0 through 9)
#$: end of string (no characters after)
```

## Part III: Adding New Columns

Based on the rubric above, we want you to add a new column to the DataFrame called *Overall Percentage* which represents the overall percentage a student has achieved in the course based on the rubric given above.

```python
#takes appropriate cols, divides by max number of points and multiplies by percentage of final grade, then by 100 to turn to percentage
sum = (((df["Class Attendance"] / 28) * 0.05) + ((df.filter(regex="^P[0-9]+$").sum(axis=1) / 1200) * 0.6) + ((df.filter(regex="^R[0-9]+$").sum(axis=1) / 500) * 0.1) + ((df["Project"] / 100) * 0.25)) * 100

df["Overall Percentage"] = sum #assigns sum to new col
df
```

## Part IV: Dropping Data

Now that you have computed the final percentages, we don't need other columns anymore. We would only like to keep the *Student*, *Class Attendance* and *Overall Percentage* column as we would use these columns to identify who qualifies to join the ministry. Write code to accomplish that below.

```python
#create new data frame with those 3 cols only
df = df[["Student", "Class Attendance", "Overall Percentage"]]
#alternative solution:
#df = df.drop(columns=df.filter(regex="^P[0-9]+$").columns.tolist() + df.filter(regex="^R[0-9]+$").columns.tolist() + ["Project"])
df
```

 ## Part V: Qualifiers for Ministry

Finally, using the thresholds and conditions given in the intro paragraph, print names of students who were qualified to work at the ministry i.e. students who have a grade that is equal to an **A- or above** AND have **attended ALL the 28 lectures**.

```python
#filter by attendance==28 and percentage >= 90
qualified_students = df[(df["Class Attendance"] == 28) & (df["Overall Percentage"] >= 90)]
qualified_students #display qualified students
```
